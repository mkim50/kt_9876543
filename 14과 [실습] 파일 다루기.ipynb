```python
text = open('news.txt', 'r', encoding='UTF8').read()
```

```python
wordList = text.split()  
```

```python
# Unique word 저장 : set으로 변환하는 이유는 중복된 값을 없애주기 때문.
worduniq = set(wordList)

# 단어와 빈도수를 담을 빈 딕셔너리 공간.
wordCount = {}

for w in worduniq:
    wordCount[w] = wordList.count(w)
```

```python
del_word = ['the','a','is','of','on','that','this','and','be','in', 'have', 'to', 'as', 'from', 'for', 'with', 'has', 'let','it', 'into', 'am', 'until', 'their', 'will', 'we', 'our', 'not', 'We', 'Let', 'I', 'The', 'were', 'at']
for w in del_word :
    if w in wordCount :
        del wordCount[w]
```

```python
# ②워드클라우드를 그립니다.
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# word cloud 그리기
wordcloud = WordCloud(width = 1000, height = 600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.show()
```
